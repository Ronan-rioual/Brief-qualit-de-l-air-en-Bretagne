```python
#Je récupère les colones qui m'intéresse

data = df[['date_ech', 'valeur', 'qualif', 'lib_zone', 'val_no2', 'val_so2', 'val_o3', 'val_pm10', 'val_pm25', 'couleur']].copy()
data.head()
```

```python
# je transforme la colone date en datetime

data['date_ech'] = pd.to_datetime(df['date_ech']).dt.date
data.head()
```

```python
# je fait la moyenne des valeur de la qualitée de l'air par date

valeur = data.groupby('date_ech')['valeur'].mean()
valeur
```

```python
# Je fait la visualisation graphique de la qualité de l'air par grande agglomérations

villes = df.lib_zone.unique()
colors = df.couleur.unique() #J'utilise ces couleurs par que il y en a 7 et que ça tombe bien avec le nombre de villes
plt.figure(figsize=(15,20))
i = 1
for color, ville in zip(colors, villes):
    df_plot = data[data['lib_zone'] == ville]
    x = np.array(df_plot['date_ech'])
    y = np.array(df_plot['valeur'])
    plt.subplot(7, 1, i)
    plt.plot(x, y, label=f'{ville}', c=color)
    plt.legend()
    i += 1
plt.show()
```

```python
df_region = df.groupby('lib_zone').mean()
df_region
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.pie(df_region, values='valeur', names=df_region.index, title="Moyenne de la qualité de l'air par zone")
fig.show()
```

```python
# moyenne de la concentration en gaz en bretagne, par gaz
no2 = data.groupby('date_ech')['val_no2'].mean()
so2 = data.groupby('date_ech')['val_so2'].mean()
o3 = data.groupby('date_ech')['val_o3'].mean()
print(no2)
print(so2)
o3
```

```python
#visualisation de la présences des différents types de gaz par villes

plt.figure(figsize=(15,20))
i = 1
for color, ville in zip(colors, villes):
    df_plot = data[data['lib_zone'] == ville]
    x = np.array(df_plot['date_ech'])
    y1 = np.array(df_plot['val_no2'])
    y2 = np.array(df_plot['val_so2'])
    y3 = np.array(df_plot['val_o3'])
    plt.subplot(7, 1, i)
    plt.plot(x, y1, label='Dioxyde d\'Azote', c='b')
    plt.plot(x, y2, label='Dioxyde de Soufre', c='g')
    plt.plot(x, y3, label='Ozone', c='r')
    plt.legend()
    plt.title(villes[i-1])
    i += 1
plt.show()
```
